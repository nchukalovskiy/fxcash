# Load latest messages from Telegram chat

```python
import os
import configparser
import pandas as pd
from datetime import datetime
import pytz
from tqdm import tqdm
import pathlib
from telethon.sync import TelegramClient
```

```python
# constants
PATH_CFG = pathlib.Path() / '../config'
PATH_SRC = pathlib.Path()
PATH_DATA = pathlib.Path() / '../data'
PATH_DATA_TG = PATH_DATA / 'telegram'
PATH_REPORTS = pathlib.Path() / '../reports'
```

```python
def date_text(date):
    return datetime.strftime(date.astimezone(tz=pytz.timezone('Europe/Moscow')), format='%Y-%m-%d %H:%M:%S')
```

```python
# —Åonfig (TELEGRAM WEB API)
config = configparser.ConfigParser()
config.read(PATH_CFG / "config.ini")

# setting configuration values
api_id = config['Telegram']['api_id']
api_hash = config['Telegram']['api_hash']
name = config['Telegram']['username']
phone = config['Telegram']['phone']
```

```python
# channel names to parse
chats = ['obmenvmsk','DollarEuroFrank']

cols = ['message_id','dt','text','sender_id','source']

for chat in chats: #chats[:1]
    
    PATH_CHAT = PATH_DATA_TG / chat 
    # find max id
    try:
        last_file_chat = max(os.listdir(PATH_CHAT))
        df = pd.read_csv(PATH_CHAT / last_file_chat)
        MAX_ID = df['message_id'].max()
    except:
        MAX_ID = 0
    # parse

    msg_lists = []

    async with TelegramClient(name, api_id, api_hash) as client:
        async for msg in client.iter_messages(chat):

            msg_list = [msg.id, 
                        date_text(msg.date),
                        msg.message,
                        msg.sender_id,
                        chat]
            print(msg_list)
            if msg.id <= MAX_ID:
                break
            msg_lists.append(msg_list)


    df_new = pd.DataFrame(msg_lists, columns=cols)
    NOW = datetime.strftime(datetime.now().astimezone(tz=pytz.timezone('Europe/Moscow')), format='%Y%m%d_%H%M%S')
    df_new.to_csv(PATH_CHAT / f'fxcash_{NOW}.csv')
```
